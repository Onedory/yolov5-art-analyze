3. 매칭안되는 이미지 삭제하기

```python
import os

dataset_path = '/content/drive/MyDrive/Data'
train_images_path = os.path.join(dataset_path, 'train/images')
train_labels_path = os.path.join(dataset_path, 'train/labels')

# 이미지 파일 목록과 라벨 파일 목록 가져오기
image_files = set(os.listdir(train_images_path))
label_files = set(os.listdir(train_labels_path))

# 이미지 파일과 라벨 파일 이름에서 확장자 제거
image_names = {os.path.splitext(f)[0] for f in image_files}
label_names = {os.path.splitext(f)[0] for f in label_files}

# 라벨 파일이 없는 이미지 파일 찾기
missing_labels = image_names - label_names

# 라벨 파일이 없는 이미지 파일 제거
for image_name in missing_labels:
    for ext in ['.jpg', '.jpeg', '.png']:  # 확장자에 맞게 수정 필요
        image_path = os.path.join(train_images_path, image_name + ext)
        if os.path.exists(image_path):
            os.remove(image_path)
            print(f'{image_path} 파일을 제거했습니다.')

# 최종 확인
num_images = len(os.listdir(train_images_path))
num_labels = len(os.listdir(train_labels_path))

print(f'Number of image files: {num_images}')
print(f'Number of label files: {num_labels}')
```

4. 매칭안되는 라벨 삭제하기

```python
import os

dataset_path = '/content/drive/MyDrive/Data'
train_images_path = os.path.join(dataset_path, 'train/images')
train_labels_path = os.path.join(dataset_path, 'train/labels')

# 이미지 파일 목록과 라벨 파일 목록 가져오기
image_files = set(os.listdir(train_images_path))
label_files = set(os.listdir(train_labels_path))

# 이미지 파일과 라벨 파일 이름에서 확장자 제거
image_names = {os.path.splitext(f)[0] for f in image_files}
label_names = {os.path.splitext(f)[0] for f in label_files}

# 이미지 파일이 없는 라벨 파일 찾기
missing_images = label_names - image_names

# 이미지 파일이 없는 라벨 파일 제거
for label_name in missing_images:
    label_path = os.path.join(train_labels_path, label_name + '.txt')
    if os.path.exists(label_path):
        os.remove(label_path)
        print(f'{label_path} 파일을 제거했습니다.')

# 최종 확인
num_images = len(os.listdir(train_images_path))
num_labels = len(os.listdir(train_labels_path))

print(f'Number of image files: {num_images}')
print(f'Number of label files: {num_labels}')
```

5. 라벨과 이미지 짝 점검

```python
import os

# 경로 설정
dataset_path = '/content/drive/MyDrive/Data'
train_images_path = os.path.join(dataset_path, 'train/images')
train_labels_path = os.path.join(dataset_path, 'train/labels')

# 이미지와 라벨 파일 목록 확인
image_files = os.listdir(train_images_path)
label_files = os.listdir(train_labels_path)

print(f"Number of image files: {len(image_files)}")
print(f"Number of label files: {len(label_files)}")

# 몇 가지 샘플 파일 확인
print("Sample image files:", image_files[:5])
print("Sample label files:", label_files[:5])

# 이미지 파일과 라벨 파일의 짝이 맞는지 확인
missing_labels = [img for img in image_files if os.path.splitext(img)[0] + '.txt' not in label_files]
missing_images = [lbl for lbl in label_files if os.path.splitext(lbl)[0] + '.jpg' not in image_files]

print(f"Images without labels: {missing_labels}")
print(f"Labels without images: {missing_images}")
```
