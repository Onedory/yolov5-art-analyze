```python
import torch
import cv2
import matplotlib.pyplot as plt

# YOLOv5 모델 로드
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/yolov5_checkpoints/yolov5_results8/weights/best.pt')  # best.pt 경로 확인

# 이미지 로드
image_path = '/content/drive/MyDrive/Data/train/images/나무_7_남_00242.jpg'  # 확인하고자 하는 이미지 경로
img = cv2.imread(image_path)

# 이미지가 제대로 로드되었는지 확인
if img is None:
    print(f"Error: 이미지 파일을 찾을 수 없거나 로드할 수 없습니다. 경로를 확인하세요: {image_path}")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR 형식으로 이미지를 로드하므로 RGB로 변환

    # 객체 감지 수행
    results = model(img)

    # 결과 시각화
    results.show()

    # 또는 Matplotlib를 사용하여 결과 시각화
    # 감지된 객체의 위치와 레이블을 이미지에 표시
    for result in results.xyxy[0].cpu().numpy():
        x1, y1, x2, y2, confidence, class_id = result[:6]
        label = model.names[int(class_id)]
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
        cv2.putText(img, f'{label} {confidence:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Matplotlib를 사용하여 결과 이미지 표시
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
import torch
import cv2
import matplotlib.pyplot as plt

# YOLOv5 모델 로드
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/yolov5_checkpoints/yolov5_results8/weights/best.pt')  # best.pt 경로 확인

# 이미지 로드
image_path = '/content/drive/MyDrive/Data/train/images/나무_7_남_00242.jpg'  # 확인하고자 하는 이미지 경로
img = cv2.imread(image_path)

# 이미지가 제대로 로드되었는지 확인
if img is None:
    print(f"Error: 이미지 파일을 찾을 수 없거나 로드할 수 없습니다. 경로를 확인하세요: {image_path}")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR 형식으로 이미지를 로드하므로 RGB로 변환

    # 객체 감지 수행
    results = model(img)

    # 경계 상자 좌표와 심리적 의미 출력
    height, width, _ = img.shape
    for result in results.xyxy[0].cpu().numpy():
        x1, y1, x2, y2, confidence, class_id = result[:6]
        label = model.names[int(class_id)]

        # 객체의 중심 좌표 계산
        cx = (x1 + x2) / 2
        cy = (y1 + y2) / 2

        # 객체가 위치한 블록 결정
        if cy < height / 3:
            vertical_pos = '상(O)'
        elif cy < 2 * height / 3:
            vertical_pos = '중앙(C)'
        else:
            vertical_pos = '하(U)'

        if cx < width / 3:
            horizontal_pos = '좌(L)'
        elif cx < 2 * width / 3:
            horizontal_pos = '중앙(C)'
        else:
            horizontal_pos = '우(R)'

        # 블록 결합하여 최종 위치
        if vertical_pos == '상(O)':
            if horizontal_pos == '좌(L)':
                position = '좌상(LU)'
            elif horizontal_pos == '우(R)':
                position = '우상(RO)'
            else:
                position = '상(O)'
        elif vertical_pos == '하(U)':
            if horizontal_pos == '좌(L)':
                position = '좌하(LO)'
            elif horizontal_pos == '우(R)':
                position = '우하(RU)'
            else:
                position = '하(U)'
        else:
            position = horizontal_pos

        # 심리적 의미 출력
        print(f"Object: {label}, Confidence: {confidence:.2f}, Position: {position}")

        # 경계 상자 그리기
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
        cv2.putText(img, f'{label} {confidence:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Matplotlib를 사용하여 결과 이미지 표시
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
