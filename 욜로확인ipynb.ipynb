```python
import torch
import cv2
import matplotlib.pyplot as plt

# YOLOv5 ëª¨ë¸ ë¡œë“œ
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/yolov5_checkpoints/yolov5_results8/weights/best.pt')  # best.pt ê²½ë¡œ í™•ì¸

# ì´ë¯¸ì§€ ë¡œë“œ
image_path = '/content/drive/MyDrive/Data/train/images/ë‚˜ë¬´_7_ë‚¨_00242.jpg'  # í™•ì¸í•˜ê³ ìž í•˜ëŠ” ì´ë¯¸ì§€ ê²½ë¡œ
img = cv2.imread(image_path)

# ì´ë¯¸ì§€ê°€ ì œëŒ€ë¡œ ë¡œë“œë˜ì—ˆëŠ”ì§€ í™•ì¸
if img is None:
    print(f"Error: ì´ë¯¸ì§€ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ê±°ë‚˜ ë¡œë“œí•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œë¥¼ í™•ì¸í•˜ì„¸ìš”: {image_path}")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCVëŠ” BGR í˜•ì‹ìœ¼ë¡œ ì´ë¯¸ì§€ë¥¼ ë¡œë“œí•˜ë¯€ë¡œ RGBë¡œ ë³€í™˜

    # ê°ì²´ ê°ì§€ ìˆ˜í–‰
    results = model(img)

    # ê²½ê³„ ìƒìž ì¢Œí‘œì™€ ì‹¬ë¦¬ì  ì˜ë¯¸ ì¶œë ¥
    height, width, _ = img.shape
    for result in results.xyxy[0].cpu().numpy():
        x1, y1, x2, y2, confidence, class_id = result[:6]
        label = model.names[int(class_id)]

        # ê°ì²´ì˜ ì¤‘ì‹¬ ì¢Œí‘œ ê³„ì‚°
        cx = (x1 + x2) / 2
        cy = (y1 + y2) / 2

        # ê°ì²´ê°€ ìœ„ì¹˜í•œ ë¸”ë¡ ê²°ì •
        if cy < height / 3:
            vertical_pos = 'ìƒ(O)'
        elif cy < 2 * height / 3:
            vertical_pos = 'ì¤‘ì•™(C)'
        else:
            vertical_pos = 'í•˜(U)'

        if cx < width / 3:
            horizontal_pos = 'ì¢Œ(L)'
        elif cx < 2 * width / 3:
            horizontal_pos = 'ì¤‘ì•™(C)'
        else:
            horizontal_pos = 'ìš°(R)'

        # ë¸”ë¡ ê²°í•©í•˜ì—¬ ìµœì¢… ìœ„ì¹˜
        if vertical_pos == 'ìƒ(O)':
            if horizontal_pos == 'ì¢Œ(L)':
                position = 'ì¢Œìƒ(LU)'
            elif horizontal_pos == 'ìš°(R)':
                position = 'ìš°ìƒ(RO)'
            else:
                position = 'ìƒ(O)'
        elif vertical_pos == 'í•˜(U)':
            if horizontal_pos == 'ì¢Œ(L)':
                position = 'ì¢Œí•˜(LO)'
            elif horizontal_pos == 'ìš°(R)':
                position = 'ìš°í•˜(RU)'
            else:
                position = 'í•˜(U)'
        else:
            position = horizontal_pos

        # ì‹¬ë¦¬ì  ì˜ë¯¸ ì¶œë ¥
        print(f"Object: {label}, Confidence: {confidence:.2f}, Position: {position}")

        # ê²½ê³„ ìƒìž ê·¸ë¦¬ê¸°
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
        cv2.putText(img, f'{label} {confidence:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Matplotlibë¥¼ ì‚¬ìš©í•˜ì—¬ ê²°ê³¼ ì´ë¯¸ì§€ í‘œì‹œ
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
